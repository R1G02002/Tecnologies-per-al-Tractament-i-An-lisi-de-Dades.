```python
# a)
IN = estudiantes.merge(notas, on="id", how="inner")
len(IN)
# Solo se muestran los id que coinciden en ambos Data Frames
```

```python
# b)
OUT = estudiantes.merge(notas, on="id", how="outer")
OUT
# Contrario que inner
```

```python
# c)
left = estudiantes.merge(notas, on="id", how="left")
left
# se dejan fuera los id del df notas q no estan en el df estudiantes
```
